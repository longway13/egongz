```python
import json
import emoji
import telegram
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from datetime import datetime
import time
from bs4 import BeautifulSoup
import random
```

```python
ID = ''
PW = ''
driver = webdriver.Chrome()
driver.get('https://everytime.kr/login')
driver.find_element(By.NAME, 'id').send_keys(ID)
driver.find_element(By.NAME, 'password').send_keys(PW)
driver.find_element(By.XPATH, '//input[@value="에브리타임 로그인"]').click()
```

```python
def text(document):
    full_text=""
    text_list=[tag.get_text() for tag in document]
    
    for text in text_list:
        full_text=full_text+text
    return full_text
```

```python
def comment(document):
    comment_list = []
    for comment in document:
        comment_dict = {
            "Author" : [],
            "Comment" : [],
            "Timestamp" : [],
            "Vote Count" : []
        }
        author = comment.find('h3', class_='medium').text.strip()
        comment_text = comment.find('p', class_='large').text.strip()
        timestamp = comment.find('time', class_='medium').text.strip()
        vote_count = comment.find('li', class_='vote').text.strip() if comment.find('li', class_='vote') else 0

        comment_dict["Author"].append(author)
        comment_dict["Comment"].append(comment_text)
        comment_dict["Timestamp"].append(timestamp)
        comment_dict["Vote Count"].append(vote_count)
        
        comment_list.append(comment_dict)
    return comment_list
```

```python
result_dict = {
        "title": [],
        "detail": [],
        "parent_comment": [],
        "child_comment": [],
        "url" : [],
        "timestamp" : []
    }
```

```python
file_path = "url_test.json"
test = []
with open(file_path, 'r') as json_file:
    data = json.load(json_file)

for url in data:
    driver.get("https://"+url)
    time.sleep(3)
    time.sleep(random.uniform(1,5))
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    
    title = text(soup.find_all('h2', class_='large'))
    detail = text(soup.find('p', class_='large'))
    parent_comments = soup.find_all('article', class_='parent')
    child_comments = soup.find_all('article', class_='child')
    
    result_dict["title"].append(title)
    result_dict["detail"].append(detail)
    result_dict["url"].append(url)
    result_dict["parent_comment"].append(comment(parent_comments))
    result_dict["child_comment"].append(comment(child_comments))
    
```
